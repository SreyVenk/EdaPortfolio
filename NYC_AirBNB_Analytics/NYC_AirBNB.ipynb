```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Read Data

```python
df = pd.read_csv("AB_NYC_2019.csv")
df.head()   #default method only shows the first 5 rows
```

# Data

```python
df.info()
df.describe()
df.isnull().sum()
```

# Cleaning Data

```python
df.drop(columns=["id", "name", "host_name", "last_review"], inplace=True)
df["reviews_per_month"] = df["reviews_per_month"].fillna(0)
```

# Univariate Analysis


1. Room Types

```python
sns.countplot(x="room_type", data=df)
plt.title("Room Type Distribution")
plt.show()
```

2. Price Distribution

```python
sns.histplot(df["price"], bins=100, kde=True)
plt.xlim(0, 500)
plt.title("Price Distribution (Capped at $500)")
plt.show()
```

3. Neighborhood Groups

```python
sns.countplot(y="neighbourhood_group", data=df)
plt.title("Listings by Borough")
plt.show()
```

# Bivariate Analysis

1. Price vs Borough

```python
sns.boxplot(x="neighbourhood_group", y="price", data=df)
plt.ylim(0, 500)
plt.title("Price by Borough")
plt.show()
```

2. Room type by Borough

```python
sns.countplot(x="neighbourhood_group", hue="room_type", data=df)
plt.title("Room Types by Borough")
plt.show()
```

3. Availability

```python
sns.histplot(df["availability_365"], bins=50)
plt.title("Availability Throughout the Year")
plt.show()
```

# Correlation Matrix

```python
plt.figure(figsize=(8, 6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()
```
